```python
import pandas as pd 
import numpy as np
```

```python
movies = pd.read_csv('tmdb_5000_movies.csv')
credits = pd.read_csv('tmdb_5000_credits.csv')
```

```python
credits_renamed = credits.rename(index=str,columns={'movie_id':'id'})
movies_df_merged = movies.merge(credits_renamed,on='id')
movies_df_merged.shape
# print(mob)
```

```python
movies_df_merged.columns
```

```python
movies_df_merged.drop(columns=['budget','keywords','title_x','title_y','cast','crew','production_countries', 'release_date', 'revenue'],axis=1,inplace=True)
```

```python
movies_df_merged.head()
```

```python
movies_df_merged.isna().sum()
```

```python
# movies_df_merged['homepage'] = movies_df_merged['homepage'].fillna('')
movies_df_merged['tagline'] = movies_df_merged['tagline'].fillna('')
```

```python
movies_df_merged.dropna(axis=0,inplace=True)
```

```python
movies_df_merged.head()
```

```python
movies_df_merged.drop('id',axis=1,inplace=True)
movies_df_merged['original_language'].nunique()
```

```python
movies_df_merged = movies_df_merged[movies_df_merged['original_language']=='en']
movies_df_merged.shape
```

```python
movies_df_merged.drop('original_language',axis=1,inplace=True)
movies_df_merged.head()
```

```python
v = movies_df_merged['vote_count']
R = movies_df_merged['vote_average']
C = movies_df_merged['vote_average'].mean()
m = movies_df_merged['vote_count'].quantile(0.70)

# Only those movies who have atleast 70% of the total votes we select only those movies
```

```python
movies_df_merged['weighted_average'] = ((R*v + C*m)/(v + m))
```

```python
movies_df_merged.head()
```

```python
movies_df_merged.drop(columns=['production_companies','status','vote_average','vote_count'],axis=1,inplace=True)
```

```python
movies_df_merged.head()
```

```python
movies_df_merged.drop(columns=['homepage','spoken_languages'],axis=1,inplace=True)
```

```python
movies_df_merged.head()
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tfv = TfidfVectorizer(min_df=3,max_features=None,strip_accents='unicode',analyzer='word',token_pattern='\w{1,}',
                      ngram_range=(1,3),stop_words='english')
```

```python
tfv_matrix = tfv.fit_transform(movies_df_merged['overview'])
tfv_matrix
```

```python
from sklearn.metrics.pairwise import sigmoid_kernel
sig = sigmoid_kernel(tfv_matrix,tfv_matrix)
```

```python
indices = pd.Series(movies_df_merged.index,index=movies_df_merged['original_title']).drop_duplicates()
```

```python
def give_rec(title,sig=sig) :
    idx = indices[title]
    sig_scores = list(enumerate(sig[idx]))
    sig_scores = sorted(sig_scores,key=lambda x: x[1],reverse=True)
    movie_ids = [i[0] for i in sig_scores]
    return movies_df_merged.iloc[movie_ids]
```

```python
give_rec('The Dark Knight Rises')
```

```python
def give_top_weight_avg():
    x = movies_df_merged.sort_values(by=['weighted_average'],ascending=False)
    return x
```

```python
def give_top_pop():
    x = movies_df_merged.sort_values(by=['popularity'],ascending=False)
    return x
```
